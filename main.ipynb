```python
import pandas as pd
```

```python
df = pd.read_csv('задание3.csv')
df.info()
```

##### Насколько заранее пользователь в среднем покупает билеты на событие?

```python
df[['timestamp', 'event_time']] = df[['timestamp', 'event_time']].apply(pd.to_datetime)
df['dif'] = df.event_time - df.timestamp
df.groupby(['user_id'], as_index=False).agg({'dif': 'median'})
```

##### Наиболее вероятный город пользователя (больше всего транзакций)

```python
df.groupby(['user_id'], as_index=False)['city_title'].agg(pd.Series.mode)
```

##### Самая популярная площадка пользователя (на какой площадке чаще покупал)

```python
df.groupby(['user_id'], as_index=False)['stage_id'].agg(pd.Series.mode)
```

##### Самая популярная категория пользователя (билеты какой категории чаще покупал)

```python
df.groupby(['user_id'], as_index=False)['category'].agg(pd.Series.mode)
```

##### Кол-во транзакций

```python
df.groupby(['user_id'], as_index=False)['transaction_id'].count()
```

##### Дата первой покупки

```python
df.groupby(['user_id'], as_index=False).agg({'timestamp': 'min'})
```

##### Дата последней покупки

```python
df.groupby(['user_id'], as_index=False).agg({'timestamp': 'max'})
```

##### Медиана цены у пользователя (Сколько стоит одна его транзакция в среднем)

```python
df.groupby(['user_id'], as_index=False).agg({'price': 'median'})
```

##### Кол-во купленных билетов

```python
df.groupby(['user_id'], as_index=False).agg({'tickets_count': 'sum'})
```

##### Продолжительность жизни пользователя (сколько он был активным и покупал /дней/)

```python
min_date_df = df.groupby(['user_id'], as_index=False).agg({'timestamp': 'min'}).timestamp
max_date_df = df.groupby(['user_id'], as_index=False).agg({'timestamp': 'max'}).timestamp
lifetime_df = pd.concat([min_date_df, max_date_df], axis=1)
lifetime_df.columns = ['min_date', 'max_date']
lifetime_df['lifetime'] = lifetime_df.max_date - lifetime_df.min_date
lifetime_df.lifetime = lifetime_df.lifetime.dt.days.replace(0, 1)
lifetime_df
```

##### Среднее кол-во билетов в одной транзакции

```python
df.groupby(['user_id'], as_index=False).agg({'tickets_count': 'median'})
```

##### Кол-во дней без транзакции (за “сегодня” взять дату ‘2022-09-20 00:00:00’)

```python
activity_df = df.groupby(['user_id'], as_index=False).agg({'timestamp': 'max'})
activity_df['dif'] = activity_df.timestamp.transform(lambda x: pd.to_datetime('2022-09-20 00:00:00') - x).dt.days
```

##### Живой или не живой пользователь (живым считаем пользователя, который совершил хотя бы одну транзакцию за последние 1.5 года)

```python
activity_df['active'] = activity_df['dif'].apply(lambda x: False if x >= 547.5 else True)
activity_df
```
