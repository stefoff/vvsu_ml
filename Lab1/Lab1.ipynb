###### Фильтрации по значениям полей

```python
rating_df = anime_df[(anime_df.rating == 8.6) | (anime_df.rating == 9)]
rating_df
```

```python
anime_df[(anime_df['members'] >= 313) & (anime_df['members'] <= 404)]
```

```python
anime_df.loc[lambda anime_df: (anime_df['members'] < 30) & (anime_df['type'] == 'OVA'), 'genre']
```

###### Сортировка отфильтрованного по рейтингу набора данных по возрастанию значений поля members

```python
rating_df.sort_values(by='members', ascending=True)
```

###### Группировка данных по среднему значению рейтинга для поля type

```python
anime_df.groupby(['type'])[['rating']].mean()
```

###### Загрузка ещё одного набора данных

```python
rating_df = pd.read_csv('data/rating.csv')
rating_df.columns
```

###### Объединение таблиц с типом присоединения inner

```python
anime_df.merge(rating_df, how = 'inner', left_on = 'anime_id', right_on = 'anime_id')
```

###### Объединение таблиц с типом присоединения left

```python
anime_df.merge(rating_df, how = 'left', left_on = 'anime_id', right_on = 'anime_id')
```

###### Объединение таблиц с типом присоединения right

```python
anime_df.merge(rating_df, how = 'right', left_on = 'anime_id', right_on = 'anime_id')
```

###### Объединение таблиц с типом присоединения outer

```python
anime_df.merge(rating_df, how = 'outer', left_on = 'anime_id', right_on = 'anime_id')
```
